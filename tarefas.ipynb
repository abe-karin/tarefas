```python
%pip install pandas
%pip install openpyxl
%pip install matplotlib
import pandas as pd

# Carregar os dados
df = pd.read_excel('arquivo.xlsx')
# Ver as primeiras linhas
df.head()
```

```python
# Preenche notas vazias com 0
df['Nota'] = df['Nota'].fillna(0)
```

```python
# Soma o tempo gasto por matéria 
tempo_por_materia = df.groupby('Matéria')['Tempo_Gasto (min)'].sum()
print(tempo_por_materia)
```

```python
import matplotlib.pyplot as plt

tempo_por_materia.plot(kind='bar', color='skyblue')
plt.title('Tempo Total de Estudo por Matéria')
plt.ylabel('Minutos')
plt.show()
```
